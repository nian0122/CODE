```python
import pandas as pd
data = {
    "姓名": ["李三", "李四", "王五", "赵六","钱七", "孙八", "周九", "吴十", "郑十一", "王十二"],
    "性别": ["男", "男", "女", "女", "男", "女", "男", "女","男", "女"],
    "身高(cm)": [172, 168, 165, 170, 178, 160, 175, 162, 180, 158],
    "体重(kg)": [65, 70, 55, 60, 75, 50,80, 53, 85, 48],
}
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

# 设置字体以支持中文
plt.rcParams["font.family"] = "SimHei"

# 1. 绘制身高分布柱状图
plt.figure(figsize=(10, 6))  # 设置绘图尺寸

# 绘制身高柱状图
plt.bar(df["姓名"], df["身高(cm)"], color="skyblue", label="身高")

# 添加标签和标题
plt.xlabel("姓名")
plt.ylabel("身高 (cm)")
plt.title("学生身高分布柱状图")
plt.legend()

# 显示图表
plt.show()

# 2. 绘制体重折线图
plt.figure(figsize=(10, 6))  # 设置绘图尺寸

# 绘制体重折线图
plt.plot(df["姓名"], df["体重(kg)"], marker="o", color="orange", label="体重")

# 添加标签和标题
plt.xlabel("姓名")
plt.ylabel("体重 (kg)")
plt.title("学生体重折线图")
plt.legend()

# 显示图表
plt.show()
```
