```python
from pydoc import describe
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

catering_sale = 'catering_sale.xls'
data = pd.read_excel(catering_sale, index_col='日期')
plt.figure()
p = data.boxplot(return_type='dict')
x = p['fliers'][0].get_xdata()  # 异常值的 x 轴坐标
y = p['fliers'][0].get_ydata()
y.sort()  # 从小到大排序
for i in range(len(x)):
    if i > 0:
        plt.annotate(y[i], xy=(x[i], y[i]), 
                    xytext=(x[i] + 0.05 - 0.8 / (y[i] - y[i-1]), y[i]), 
                    arrowprops=dict(arrowstyle='-', connectionstyle='arc3', color='gray'), 
                    fontsize=8, color='red')
    else:
        plt.annotate(y[i], xy=(x[i], y[i]), 
                    xytext=(x[i] + 0.05, y[i]), 
                    arrowprops=dict(arrowstyle='-', connectionstyle='arc3', color='gray'), 
                    fontsize=8, color='red')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
catering_sale = 'catering_fish_congee.xls'  # 文件路径
data = pd.read_excel(catering_sale, names=['date', 'sale'])  # 读取数据，并指定列名为 "date" 和 "sale"
bins = [0, 500, 1000, 1500, 2000, 2500, 3000, 3500, 4000]
labels = ['(0, 500)', '(500, 1000)', '(1000, 1500)', '(1500, 2000)', '(2000, 2500)', '(2500, 3000)', '(3000, 3500)', '(3500, 4000)']
data['sale 分层'] = pd.cut(data['sale'], bins=bins, labels=labels)
aggResult = data.groupby('sale 分层').agg({'sale': 'count'})
paggResult = round(aggResult / aggResult.sum() , 2,)*100

plt.figure(figsize=(8, 4))
paggResult['sale'].plot(kind='bar', width=0.8, fontsize=10)

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.title('季度销售额频率分布直方图', fontsize=20)
plt.show()
catering_dish_profit = 'catering_dish_profit.xls'  # 文件路径
data = pd.read_excel(catering_dish_profit)  # 读取数据，并指定 "日期" 列为索引
plt.figure(figsize=(8, 6))
plt.pie(data['盈利'], labels=data['菜品名'])
plt.title('菜品销售量分布（饼图）')

plt.figure(figsize=(8, 4))
x = data['菜品名']
y = data['盈利']
plt.bar(x, y)
plt.rcParams['font.sans-serif'] = ['SimHei']

plt.xlabel('菜品名')
plt.ylabel('销量')
plt.title('菜品销售量分布（条形图）')

plt.show()
```
