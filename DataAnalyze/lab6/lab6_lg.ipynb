```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#数据读入
from sklearn.datasets import load_breast_cancer
breast_cancer=load_breast_cancer()
```

```python
#显示数据集字典的键
print(breast_cancer.keys())
```

```python
print(breast_cancer.DESCR)
```

```python
#数据集的特征名称
print(breast_cancer.feature_names)
```

```python
#数据形状
print(breast_cancer.data.shape)
```

```python
#将数据（data）与回归目标(target)转换为数据框类型
X=pd.DataFrame(breast_cancer.data,columns=breast_cancer.feature_names)
y=pd.DataFrame(breast_cancer.target,columns=['class'])
```

```python
#合并数据框
df=pd.concat([X,y],axis=1)
df
```

```python
#查看数据基本信息
df.info()
```

```python
#查看描述新统计信息
df.describe()
```

```python
#划分训练集和测试集
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               test_size=0.3,random_state=42)
X_test
```

```python
#模型训练
model=LogisticRegression(C=1.0,
                         class_weight=None,
                         dual=False,
                         fit_intercept=True,
                         intercept_scaling=1,
                         max_iter=100,
                         multi_class='ovr',
                         n_jobs=1,
                         penalty='l2',
                         random_state=None,
                         solver='liblinear',
                         tol=0.0001,
                         verbose=0,
                         warm_start=False)
model.fit(X_train,y_train)
```

```python
#默认参数模型预测结果y_pred
y_pred=model.predict(X_test)
y_pred
```

```python
#混淆矩阵（分类算法的重要评估指标）
matrix=metrics.confusion_matrix(y_test,y_pred)
matrix
```

```python
#分类评价指标——准确度（Accuracy）
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
 
#分类评价指标——精度（Precision）
print("Precision:",metrics.precision_score(y_test,y_pred))
```

```python
#以C、penalty参数和值设置字典列表param_grid。
#设置cv参数值为5
param_grid={'C':[0.001,0.01,0.1,1,10,20,50,100],
            'penalty':["l1","l2"]}
n_folds=5
```

```python
#调用GridSearchCV函数，进行5折交叉验证，对估计器LogisticRegression()的指定参数值param_grid进行详尽搜索，得到最终的最优模型参数
from sklearn.model_selection import GridSearchCV
estimator=GridSearchCV(LogisticRegression(solver='liblinear'),param_grid,cv=n_folds)
estimator.fit(X_train,y_train)
```

```python
estimator.best_estimator_
```

```python
#调参后的模型训练
model1=LogisticRegression(C=50,
                         class_weight=None,
                         dual=False,
                         fit_intercept=True,
                         intercept_scaling=1,
                         max_iter=100,
                         multi_class='ovr',
                         n_jobs=1,
                         penalty='l1',
                         random_state=None,
                         solver='liblinear',
                         tol=0.0001,
                         verbose=0,
                         warm_start=False)
model1.fit(X_train,y_train)
```

```python
#调参后的模型预测
y_pred1=model1.predict(X_test)
y_pred1
```

```python
#调参后的模型混淆矩阵结果
matrix1=metrics.confusion_matrix(y_test,y_pred1)
matrix1
```

```python
#调参后的额模型准确度、精度分类指标评价结果
print("Accuracy1:",metrics.accuracy_score(y_test,y_pred1))
print("Precision1:",metrics.precision_score(y_test,y_pred1))
```
